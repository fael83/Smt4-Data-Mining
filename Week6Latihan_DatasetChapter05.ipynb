```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
from google.colab import drive
drive.mount('/content/drive')

df_chap = pd.read_csv('/content/drive/MyDrive/Dataset Tugas Data Mining/Chapter05DataSet Tugas Minggu 6.csv')
df_chap.head()
```

```python
df_bool = df_chap.astype(bool)
df_bool.head()
```

```python
frequent_items = apriori(df_bool, min_support=0.1, use_colnames=True)
frequent_items
```

```python
rules = association_rules(frequent_items, metric="confidence", min_threshold=0.5)
rules
```
